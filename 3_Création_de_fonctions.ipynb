**Un module s'utilise de la manière suivante** :
- `module(variable)`   où `variable` est l'élément sur lequel on applique le module.

**Voici les modules standards de python** :
- `type`  : Renvoi le type de la variable passée en paramètre,
- `print` : Affiche la variable passée en argument,
- `int`   : Transforme un nombre réel en entier naturel,
- `float` : Transforme un nombre quelconque en nombre réel,
- `str`   : Transforme la variable passée en entrée en chaine de caractère,
- `bool`  : Renvoi True ou False en fonction de la variable passée en paramètre,
- `list`  : Renvoi une list en fonction de l'itérable passée en entrée,
- `len`   : Renvoi la taille de la liste entrée en paramètre,
- `dir`   : Renvoi la liste des méthodes de l'objet entré en paramètre.
- `input` : Crée une zone de saisie de text.
- `range` : Renvoi une liste de nombre entier de 0 à `n`, où `n` est la variable passée en paramètre.

**Opérateurs logiques** :

- `==` : Renvoi True si les deux éléments comparés sont égaux --> exemple : `3+3 == 6`


- `!=` : Renvoi True si les deux éléments comparés sont différent --> exemple : 3 == 6


- `<` (ou `<=`) : Renvoi `True` si l'élément de gauche est plus petit (ou égale) à celui de droite --> exemple : `3 < 6`


- `>` (ou `>=`) : Renvoi `False` si l'élément de gauche est plus grand (ou égale) à celui de droite --> exemple : `3 > 6`

## 1.3 Créez une fonction nommée `display_nx_some_text` qui prend en paramétre `n` et `some_text` et qui affiche `n` fois la chaine de caractère `some_text`.

```python
def display_nx_some_text(n, some_text):
    for i in range(n):
        print(some_text)

display_nx_some_text(5, "Hello guys")
```

## 1.4 Créez une fonction nommée `display_input_user` qui demande à l'utilisateur de taper du text puis affiche un à un tous les mots tapés par l'utilisateur.

```python
def display_input_user():
    user_input = input("Tape ton texte : ")
    words = user_input.split()
    
    for i, word in enumerate(words):
        if i < 20:
            print(word)
        else:
            break

display_input_user()

```

## 1.5 Créez une fonction nommée `get_name_user` qui prend en paramètre un booléen.
- Si ce booléen est `True`, la fonction demande à l'utilisateur son nom et retourne cette valeur.

- Si ce boolen est `False`, la fonction retourn la chaine de caractère `John Doe`.

```python
def get_name_user(param: bool):
    if param:
        user_name = input("Écris ton nom: ")
        return user_name
    else:
        return "John Doe"

user_name = get_name_user(True)
print(user_name)

default_name = get_name_user(False)
print(default_name)
```

## 1.7 Créez une fonction nommée `multiplicate_by_two`, qui prend deux arguments :
- Le premier : un nombre.
- Le second  : un booléen par défaut sur True.

Si le booléen est True, la fonction affiche avec `print` la multiplication par 2.

Sinon la fonction retourn avec `return` la multiplication par 2.

```python
def multiplicate_by_two(number, print_result=True):
    result = number * 2
    if print_result:
        print(f"Le résultat est : {result}")
    else:
        return result

result_without_print = multiplicate_by_two(5, False)
print(result_without_print)
```
