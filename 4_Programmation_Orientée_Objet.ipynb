# 4. Programmation Orientée Objet (POO)

___

La programmation orientée objet en Python est une méthode de programmation qui utilise des objets et des classes pour structurer le code.

Les objets sont des instances de classes, qui définissent les propriétés et les méthodes de l'objet.

Les classes peuvent hériter les propriétés et les méthodes d'autres classes, ce qui permet de créer des relations de parenté entre les objets.

En utilisant la programmation orientée objet, on peut créer des programmes plus organisés et plus faciles à maintenir.

- `module(variable)`   où variable est l'élément sur lequel on applique notre module.

**Voici les modules standards de python** :
- `type`  : Renvoi le type de la variable entrée en paramètre
- `print` : Affiche la variable passée en argument,
- `int`   : Transforme un nombre réel en entier naturel,
- `float` : Transforme un nombre quelconque en nombre réel,

**Opérateurs logiques** :

- `==` : Renvoie True si les deux éléments comparés sont égaux --> exemple : 3+3 == 6


- `!=` : Renvoie True si les deux éléments comparés sont différent --> exemple : 3 == 6


- `<` (ou `<=`) : Renvoie True si l'éléments de gauche est plus petit (ou égale) à celui de droite --> exemple : 3 < 6


- `>` (ou `>=`) : Renvoie False si l'éléments de gauche est plus grand (ou égale) à celui de droite --> exemple : 3 > 6

# 1 Définition d'une classe


Une classe est un modèle pour créer des objets (instances), qui ont des propriétés (attributs) et des comportements (méthodes).

Les attributs sont des variables de classe,
tandis que les méthodes sont des fonctions de classe.

Lorsqu'un objet est créé à partir d'une classe ont dit qu'il a été instancié.

```python
class NomDeClasse:
  def __init__(self, param1, param2):
    self.a =0
    self.test1 = param1
    self.test2 = param2
```

```python
# Déclaration d'une classe, paramètre 1 et 2 seront

class NomDeLaClasse:
  # Définition des attributs
  def __init__(self, parametre1, parametre2):
    self.parametre1 = parametre1
    self.parametre2 = parametre2

  # Définition d'une méthode
  def nom_de_la_methode(self, p1, p2):
    pass


# Instance de classe
objet = NomDeLaClasse('parametre1', 'parametre2')



#Héritage :
class ClasseFille(NomDeLaClasse):
  def __init__(self, parametre1, parametre2, parametre3):
    NomDeLaClasse.__init__(self, parametre1, parametre2)
    self.parametre3 = parametre3
```

# 1.1 Créez une classe `Voiture` qui aura en attribut :
- Le nom de la voiture
- Le modèle de la voiture
- La vitesse maximale de cette voiture
- Son kilométrage
- Son contrôle technique (True ou False)

Instanciez ensuite un objet `voiture1` avec les valeurs suivantes :    
- 'Renault'
- 'Clio 3'
- 260

- True

```python
class Voiture:
    def __init__(self, name, model, vit, km, ct):
        self.name = name 
        self.model = model 
        self.vit = vit 
        self.km = km 
        self.ct = ct 

voiture1 = Voiture('Opel', 'Astra GTC', 360, 25000, True)

print("Nom de la voiture :", voiture1.name)
print("Modèle de la voiture :", voiture1.model)
print("Vitesse maximale :", voiture1.vit)
print("Kilométrage :", voiture1.km)
print("Contrôle technique :", voiture1.ct)
```

# 1.2 Ajoutez une méthode `Trajet` qui prend trois paramètres :    
- Destination
- Arrivée
- Distance

Cette méthode crée deux attributs :
- Arrivée
- Distance

Cette méthode ajoute la distance parcourue au kilométrage de la voiture.


Si le kilométrage est supérieur à 500 km, l'attribut 'contrôle technique' doit passer sur False.

```python
class Voiture:
    def __init__(self, name, model, vit, km, ct):
        self.name = name
        self.model = model
        self.vit = vit
        self.km = km
        self.ct = ct

    def Trajet(self, destination, arrivee, distance):
        self.arrivee = arrivee
        self.distance = distance
        self.km += distance
        if self.km > 500:
            self.ct = False

voiture1 = Voiture('Opel', 'Astra GTC', 360, 25000, True)

voiture1.Trajet('Paris', 'Paris', 600)

print("Nom de la voiture :", voiture1.name)
print("Modèle de la voiture :", voiture1.model)
print("Vitesse maximale :", voiture1.vit)
print("Kilométrage :", voiture1.km)
print("Contrôle technique :", voiture1.ct)
print("Arrivée du dernier trajet :", voiture1.arrivee)
print("Distance du dernier trajet :", voiture1.distance)
```

# 2. Créez une classe Humain.

La classe doit avoir les attributs suivants :    
- Nom
- Prénom
- Age
- Genre
- Diplomé
- Solde
- date de naissance
- majorité


La classe doit avoir les méthode suivantes :
- banque : qui modifie l'attribut solde
- meeting : qui prend en argument une instance Humain et affiche le nom des deux Humains.
- majeure : qui vérifie si l'humain est majeur en fonction de la date du jour.



```python
from datetime import date

class Humain:
    def __init__(self, nom, prenom, age, genre, diplome, solde, date_naissance):
        self.nom = nom
        self.prenom = prenom
        self.age = age
        self.genre = genre
        self.diplome = diplome
        self.solde = solde
        self.date_naissance = date_naissance
        self.majorite = self.majeure()

    def banque(self, montant):
        self.solde += montant

    def meeting(self, autre_humain):
        print(f"{self.prenom} & {autre_humain.prenom}")

    def majeure(self):
        if self.age >= 18:
            return True
        return False

humain1 = Humain("Gass", "Gwen", 37, "Non binaire", "Bac+3", 2000, date(1986, 2, 21))
humain2 = Humain("Gass", "Gwendo", 26, "Femme", "Bac+5", 1200, date(1997, 1, 17))

print("Solde avant la transaction :", humain1.solde)
humain1.banque(500)
print("Solde après la transaction :", humain1.solde)

humain1.meeting(humain2)

if humain1.majorite and humain2.majorite:
    print(f"{humain1.prenom} et {humain2.prenom} sont majeurs.")
else:
    print(f"Certains des humains ne sont pas encore majeurs.")
```
